```python
from langchain.tools import Tool
from langchain.utilities import GoogleSearchAPIWrapper

search = GoogleSearchAPIWrapper()

def top5_results(query):
    return search.results(query=query, num_results=5)

tool = Tool(
    name="Sample",
    description="Search Google for recent results.",
    func=top5_results,
)
```

```python
from bs4 import BeautifulSoup
import requests

def getHtmlContentFromUrl(url):
  response = requests.get(url, timeout=10)
  html_content = response.text
  soup = BeautifulSoup(html_content, 'lxml')
  for script in soup.find_all('script'):
    script.extract()
  list_content = soup.find_all('p')
  joined_content = ''
  for content in list_content:
    joined_content += content.text
  return joined_content
```

```python
url = 'https://www.hyundai.co.kr/story/CONT0000000000018191'  # Replace with the URL of the webpage you want to extract from

print(getHtmlContentFromUrl(url=url))
```
