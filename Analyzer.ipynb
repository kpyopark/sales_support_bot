```python
import vertexai

PROJECT_ID = os.getenv("PROJECT_NUMBER")  # @param {type:"string"}
vertexai.init(project=PROJECT_ID, location="us-central1")
```

```python
from langchain.llms import VertexAI


llm = VertexAI(
    model_name="text-bison@latest",
    max_output_tokens=1024,
    temperature=0,
    top_p=0.8,
    top_k=40,
)
```

```python
def checkNanValueCellCountInDf(df):
    return df.isnull().sum().sum()

def checkValidValueCellsInDf(df):
    return df.count().sum()

for df in df_list:
    if checkNanValueCellCountInDf(df) == 0:
        print(df)
        continue
    print("################## BAD DATA ##################")
    print(df)
    print("################## BAD DATA ##################")
```

```python

prompt_template_valueable = """As you are CEO in the company - {company} - , you should determine wheter the given context is valuable or not for creating new business agenda. Your answer should be decimal points from 0(not valuable) to 100(most valuable).

context:
{context}"""

# this prompot isn't well suited.

prompt_template_agenda = """As you are CEO in a company, you should determine wheter the given context is valuable or not. If the give context has values, answer how does the future business direction appear?. If you can't find any directions, just answer 'no'.

context:
{context}"""

def makeAgendas(df_list, company_name):
  list_of_data = []
  list_of_agenda = []
  for df in df_list:
    list_of_data.append(df.to_string())
    if checkNanValueCellCountInDf(df) == 0:
      response = llm.predict(prompt_template_agenda.format(company=company_name, context=df.to_string()))
      print(df.to_string())
      print(response)
      list_of_agenda.append(response)
    else:
      list_of_agenda.append("")
  return list_of_data, list_of_agenda

```

```python
list_of_data, list_of_agenda = makeAgendas(df_list=df_list, company_name="이마트")
```

```python
df_result = pd.DataFrame({"data": list_of_data, "agenda": list_of_agenda})
```

```python
prompt_template_cfo_role = """As the CFO role, please summarize the Business Goals and Actions to be taken next year for the given sentence. Define larger categories for Business Goals. Provide the response in the format of Category | Business Goal | Action.

context:
{context}"""

prompt_template_cto_role = """As the CTO role, please summarize the Business Goals and Actions to be taken next year for the given sentence. Define larger categories for Business Goals. Provide the response in the format of Category | Business Goal | Action.

context:
{context}"""
```

```python
def makeBusinessGoalAsCfo(business_agendas):
  list_of_business_goal = []
  agendas = ""
  for agenda in business_agendas:
    agendas += agenda
    if len(agendas) > 7000:
      response = llm.predict(prompt_template_cfo_role.format(context=agendas))
      print(response)
      list_of_business_goal.append(response)
      agendas = ""
  if len(agendas) > 0:
    response = llm.predict(prompt_template_cfo_role.format(context=agendas))
    print(response)
    list_of_business_goal.append(response)
  return list_of_business_goal
```

```python
list_of_business_goal_as_cfo = makeBusinessGoalAsCfo(df_result["agenda"])
```
